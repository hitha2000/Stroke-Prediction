# Stroke Prediction 

# Exploratory Data Analysis

```python
# importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
```

```python
# Read data
stroke_prediction_df = pd.read_csv("stroke-prediction-data.csv")
stroke_prediction_df.head()
```

```python
# inspecting the dataset structure

stroke_prediction_df.info()
```

```python
# Dimensions of the dataset

stroke_prediction_df.shape
```

```python
# Describe the data

round(stroke_prediction_df.describe(include='all'),2)
```

```python
# Drop "id" col from DF

stroke_prediction_df.drop('id', axis=1, inplace=True)
```

```python
# Remove duplicates

stroke_prediction_df.drop_duplicates(inplace=True)
```

```python
# Identify the missing values

stroke_prediction_df.isna().sum()
```

```python
# Calculating the percentage of missing values in the dataset

((stroke_prediction_df.isna().sum())/ len(stroke_prediction_df)) * 100
```

```python
# Filling the missing values of bmi with median 

stroke_prediction_df.fillna(stroke_prediction_df['bmi'].median(skipna=True),inplace=True)
```

```python
# Analysis and Transformation of Outliers.

# The features age, bmi and avg_glucose_level are numerical values and are more prone to have outliers, while the other features 
# are either binary or nominal. 

# Boxplot of “age”,“bmi”, and “avg_glucose_level” features to showcase outliers.

stroke_prediction_df.boxplot(column =['age', 'bmi','avg_glucose_level'])
plt.show()
```

```python
# Transforming “bmi” column using loge (X + 1) formula to make the “bmi” values follow the normal distribution
stroke_prediction_df.loc[:,'bmi'] = np.log(stroke_prediction_df['bmi']+1)
```

```python
stroke_prediction_df.loc[:,'avg_glucose_level'] = np.log(stroke_prediction_df['avg_glucose_level']+1)
```

# Visualization

```python
numeric_columns = ['age', 'bmi', 'avg_glucose_level']
categorical_columns = ['gender', 'hypertension', 'heart_disease', 'ever_married', 'work_type', 'Residence_type', 'smoking_status', 'stroke']
```

```python
i = 0
fig, ax = plt.subplots(3, 3, figsize=(15, 8))
plt.subplots_adjust(hspace = 0.5)
for num_col in numeric_columns :
    sns.kdeplot(x=num_col, hue='stroke', data=stroke_prediction_df, multiple='stack', ax=ax[i,0])
    sns.boxplot(x=num_col, data=stroke_prediction_df, ax=ax[i, 1])
    sns.scatterplot(x=num_col, y='stroke', data=stroke_prediction_df, ax=ax[i, 2])
    i+=1
plt.show()
```

```python
i=0
while i<8 :
    
    # Left AX
    fig = plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.title(categorical_columns[i], size=20, weight='bold', color='navy')
    ax = sns.countplot(x=categorical_columns[i], data=stroke_prediction_df)
    ax.bar_label(ax.containers[0])
    ax.tick_params(axis='x', rotation=300)
    i+=1
    
    # Right AX
    plt.subplot(1, 2, 2)
    plt.title(categorical_columns[i], size=20, weight='bold', color='navy')
    ax = sns.countplot(x=categorical_columns[i], data=stroke_prediction_df)
    ax.bar_label(ax.containers[0])
    ax.tick_params(axis='x', rotation=300)
    i+=1
    plt.show()
```

```python
x = stroke_prediction_df['stroke'].value_counts()


explode = [0, 0.15]
labels = ['Stroke=0', 'Stroke=1']
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(aspect="equal"))

plt.pie(x, explode=explode, shadow=True, autopct='%1.1f%%', labels=labels, textprops=dict(color="w", weight='bold', size=15))
plt.legend()
plt.show()
```

```python
# Encode the categorical variables with integer values

label_encoder = LabelEncoder()
stroke_prediction_df[stroke_prediction_df.select_dtypes(include=['object']).columns] = stroke_prediction_df.select_dtypes(include=['object']).apply(lambda x: label_encoder.fit_transform(x))
stroke_prediction_df
```

```python
# Feature Scaling using Min-Max Normalization

X_value = stroke_prediction_df.drop(columns ='stroke')
y = stroke_prediction_df.stroke

scaler = MinMaxScaler().fit_transform(X_value)
X = pd.DataFrame(scaler, columns=X_value.columns)
X.describe()
```

# Machine Learning Modeling

```python
# Splitting the data into training and test data

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.25)
```

```python
# training the dataset using Decision Tree model

tree = DecisionTreeClassifier(criterion = 'entropy',splitter='random',max_depth=4)
tree.fit(X_train,y_train)
y_pred = tree.predict(X_test)

print('accuracy : ',accuracy_score(y_test,y_pred))
```

```python
# Plot the confusion matrix


cm = confusion_matrix(y_test, y_pred)
ax = plt.subplot()
ax = sns.heatmap(cm, annot=True, fmt='', cmap="Oranges")
ax.set_xlabel('Prediced labels')
ax.set_ylabel('True labels')
ax.set_title('Confusion Matrix')
ax.xaxis.set_ticklabels(['Dont Had Stroke', 'Had Stroke'])
ax.yaxis.set_ticklabels(['Dont Had Stroke', 'Had Stroke']) 
plt.show()
```

```python
# Classification report

print('Classification Report : \n',classification_report(y_test,y_pred, zero_division=0))
```
